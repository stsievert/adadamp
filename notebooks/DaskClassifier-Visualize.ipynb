# Visualization for Data from DaskClassifier

```python
import json
import matplotlib.pyplot as plt
```

```python
with open('./stats_10epoch.json', 'r') as fp:
    data = json.load(fp)
```

```python
meta = data['meta']
batch = data['batch']
print('Meta Keys:\t{}\nBatch Keys:\t{}'.format(list(meta.keys()), list(batch[0].keys())))
```

```python
def plot(y, x_label, y_label, title):
    plt.plot(y)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
```

```python
# plot meta info
score_times = meta['score_times']
plot(score_times, 'Score Call', 'Time (s)', 'Time to Compute Score')
print('Meta Info:')
print('\tNum Updates:', meta['n_updates'])
print('\tNum Data:', meta['n_data'])
print('\tScore Calls:', meta['score_calls'])
print('\tFit Calls:', meta['fit_calls'])
print()
```

```python
# epoch time over time
epoch_times = [pt['epoch_time'] for pt in batch]
plot(epoch_times, 'Epochs', 'Time (s)', 'Time Per Epoch')
```

```python
# score time over time
score_times = meta['score_times']
plot(score_times, 'Score Call', 'Time (s)', 'Time Per Score')
```

```python
# acc over time
accs = [pt['acc'] for pt in batch]
plot(accs, 'Batch', 'Accuracy', 'Accuracy for Batch')
```

```python
# acc over time
accs = [pt['acc'] for pt in batch][:100]
plot(accs, 'Batch', 'Accuracy', 'Accuracy for Batch (truncated)')
```

```python
# Loss over time
accs = [pt['loss'] for pt in batch]
plot(accs, 'Batch', 'Loss', 'Loss for Batch')
```

```python
# Loss over time
accs = [pt['loss'] for pt in batch][:100]
plot(accs, 'Batch', 'Loss', 'Loss for Batch')
```
